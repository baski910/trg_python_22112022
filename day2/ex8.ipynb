```python
# working with images using numpy
import imageio
import numpy as np
import matplotlib.pyplot as plt

pic = imageio.imread('demo_2.jpg')

print(pic.shape)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(pic)
plt.show()
```

```python
# render the image in black and white
g = lambda img: np.dot(img[...,:3],[0.299,0.587,0.142])
gray = g(pic)
plt.figure(figsize=(12,10))
plt.imshow(gray,plt.get_cmap(name='gray'))
plt.show()
```
